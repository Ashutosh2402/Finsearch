```python
# Finsearch
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
from scipy.stats import norm
import pandas as pd
from matplotlib import pyplot as plt
niftyhd=pd.read_csv('/content/drive/MyDrive/FinsearchE14/NIFTY
50-26-08-2022-to-26-08-2023.csv')
niftyhd=niftyhd.rename(columns={'Close ': 'Close'})
# Removing an extra space
nhdClose=niftyhd.Close
# Getting daily closing data for the past year to compute sdandard deviation
m=nhdClose.mean()
s=nhdClose.std()
rsd=s/m
#rsd is the Relative standard deviation and we will be using it as a measure of the
volatitlity of the stock
'''
The location of the data used in the might be different and may result in
an error Please verify the location of the data files before running the code
'''
oc31aug=pd.read_csv('/content/drive/MyDrive/FinsearchE14/optionchain
NIFTY31Aug2023.c sv')
oc07sep=pd.read_csv('/content/drive/MyDrive/FinsearchE14/optionchain
NIFTY07Sep2023.c sv')
cp=19265.80
# This is the current price of the stock on the day whose option prices we are analyzing.
rfr=0.07206 #risk free intrest rate is assumed to be 7.206% as India 10 Years
Government Bond has a 7.206% yield
# Using the black scholes formula to define the function
def BS_CALL(S, K, T, r, sigma):
   d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
   d2 = d1 - sigma * np.sqrt(T)
   return S * norm.cdf(d1) - K * np.exp(-r*T)* norm.cdf(d2)
def BS_PUT(S, K, T, r, sigma):
   d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
   d2 = d1 - sigma* np.sqrt(T)
   return K*np.exp(-r*T)*norm.cdf(-d2) - S*norm.cdf(-d1)
c31aug=[]
p31aug=[]
c07sep=[]
p07sep=[]
for i in oc31aug.STRIKE:
   c31aug.append(BS_CALL(cp,float(i),6/365,rfr,rsd))
#6/365 refers to time in years
   p31aug.append(BS_PUT(cp,float(i),6/365,rfr,rsd))
#6/365 refers to time in years
for i in oc07sep.STRIKE:
   c07sep.append(BS_CALL(cp,float(i),13/365,rfr,rsd))
#13/365 refers to time in years
   p07sep.append(BS_PUT(cp,float(i),13/365,rfr,rsd))
#13/365 refers to time in years
a=plt.axes()
a.scatter(oc31aug.STRIKE,c31aug,c='red',s=5)
a.scatter(oc31aug.STRIKE,oc31aug.LTPC,c='blue',s=5)
a.set_xlabel("Strike Price")
a.set_ylabel("Option Premium")
a.legend(["Black Scholes","Real Data"])
plt.title("Call,Timeframe:25Aug-31Aug")
a=plt.axes()
a.scatter(oc31aug.LTPC,c31aug,c='red',s=5)
a.set_xlabel("Real Data")
a.set_ylabel("Black Scholes")
plt.title("Call,Timeframe:25Aug-31Aug,BS vs Real")
b=plt.axes()
b.scatter(oc31aug.STRIKE,p31aug,c='red',s=5)
b.scatter(oc31aug.STRIKE,oc31aug.LTPP,c='blue',s=5)
b.set_xlabel("Strike Price")
b.set_ylabel("Option Premium")
b.legend(["Black Scholes","Real Data"])
plt.title("Put,Timeframe:25Aug-31Aug")
b=plt.axes()
b.scatter(oc31aug.LTPP,p31aug,c='red',s=5)
b.set_xlabel("Real Data")
b.set_ylabel("Black Scholes")
plt.title("Put,Timeframe:25Aug-31Aug,BS vs Real")
c=plt.axes()
c.scatter(oc07sep.STRIKE,c07sep,c='red',s=5)
c.scatter(oc07sep.STRIKE,oc07sep.LTPC,c='blue',s=5)
c.set_xlabel("Strike Price")
c.set_ylabel("Option Premium")
c.legend(["Black Scholes","Real Data"])
plt.title("Call,Timeframe:25Aug-07Sep")
c=plt.axes()
c.scatter(oc07sep.LTPC,c07sep,c='red',s=5)
c.set_xlabel("Real Data")
c.set_ylabel("Black Scholes")
plt.title("Call,Timeframe:25Aug-07Sep,BS vs Real")
d=plt.axes()
d.scatter(oc07sep.STRIKE,p07sep,c='red',s=5)
d.scatter(oc07sep.STRIKE,oc07sep.LTPP,c='blue',s=5)
d.set_xlabel("Strike Price")
d.set_ylabel("Option Premium")
d.legend(["Black Scholes","Real Data"])
plt.title("Put,Timeframe:25Aug-07Sep")
d=plt.axes()
d.scatter(oc07sep.LTPP,p07sep,c='red',s=5)
d.set_xlabel("Real Data")
d.set_ylabel("Black Scholes")
plt.title("Put,Timeframe:25Aug-07Sep,BS vs Real")
#Finding correlation coefficient for real vs live data
cc31augc=np.corrcoef(oc31aug.LTPC,c31aug)
print("The Correlation coefficient for BS vs REAL for Call
from 25Aug-31Aug:",end=' ')
print(cc31augc[0,1])
cc31augp=np.corrcoef(oc31aug.LTPP,p31aug)
print("The Correlation coefficient for BS vs REAL for Put from
25Aug-31Aug:",end=' ')
print(cc31augp[0,1])
cc07sepc=np.corrcoef(oc07sep.LTPC,c07sep)
print("The Correlation coefficient for BS vs REAL for Call
from 25Aug-07Sep:",end=' ')
print(cc07sepc[0,1])
cc07sepp=np.corrcoef(oc07sep.LTPP,p07sep)
print("The Correlation coefficient for BS vs REAL for Put from
25Aug-07Sep:",end=' ')
print(cc07sepp[0,1])
```
